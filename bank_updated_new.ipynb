```python
# Histogram after mean imputation

# Create a copy of x_new to perform mean imputation
x_new_mean = x_new.copy()

# Iterate over each column in x_new_mean for outlier detection and mean imputation
for i in x_new_mean.columns:
    # Calculate the first quartile (Q1), third quartile (Q3), mean, and interquartile range (IQR)
    q1 = np.quantile(x_new_mean[i], 0.25)
    q3 = np.quantile(x_new_mean[i], 0.75)
    mean = np.mean(x_new_mean[i])
    iqr = q3 - q1
    
    # Calculate upper and lower bounds for outlier detection using IQR method
    upper_bound = q3 + 1.5 * iqr
    lower_bound = q1 - 1.5 * iqr
    
    # Replace outliers with mean value in the current column i of x_new_mean
    x_new_mean[i] = np.where((x_new_mean[i] > upper_bound) | (x_new_mean[i] < lower_bound), mean, x_new_mean[i])

# Iterate over each column in x_new_mean and plot a histogram with KDE (Kernel Density Estimation)
for i in x_new_mean.columns:
    # Create a new figure for each histogram with specified size
    plt.figure(figsize=(8, 6))
    
    # Use seaborn's histplot with KDE to create the histogram for the current column i in x_new_mean
    sns.histplot(x_new_mean[i], kde=True)
    
    # Display the plot
    plt.show()
```

```python
# Bagging for Decision tree

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

dt= DecisionTreeClassifier()#DecisionTreeClassifier() is the base classifier used within the bagging ensemble.

bag_clf= BaggingClassifier(estimator=dt,
                           n_estimators= 100,
                           max_samples= 0.6,
                           max_features= 0.7)# BaggingClassifier is configured to use 100 estimators (Decision Trees), 
#each trained on 60% of the samples (max_samples=0.6) and 70% of the features (max_features=0.7).

training_score=[]
testing_score=[]
testing_precision=[]
testing_recall=[]
testing_f1=[]


for i in range(1,100):
  x_train,x_test,y_train,y_test= train_test_split(x_resample, y_resample, test_size= 0.3, random_state= i)
  bag_clf.fit(x_train, y_train)
  y_pred_train= bag_clf.predict(x_train)
  y_pred_test= bag_clf.predict(x_test)
  training_score.append(accuracy_score(y_train, y_pred_train))
  testing_score.append(accuracy_score(y_test, y_pred_test))
  testing_precision.append(precision_score(y_test, y_pred_test))
  testing_recall.append(recall_score(y_test, y_pred_test))
  testing_f1.append(f1_score(y_test, y_pred_test))
print("Accuracy score of train:", np.mean(training_score).round(2))
print("Accuracy score of test:", np.mean(testing_score).round(2))
print("Precision score of test:", np.mean(testing_precision).round(2))
print("Recall score of test:", np.mean(testing_recall).round(2))
print("F1 score of test:", np.mean(testing_f1).round(2))
```
